```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='cut', y='price', data=df, palette="Set2")
plt.title('Distribusi Harga Berlian Berdasarkan Potongan', fontsize=14)
plt.xlabel('Cut', fontsize=12)
plt.ylabel('Price (USD)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=30, kde=True, color='purple', edgecolor='black')
plt.title('Distribusi Harga Berlian', fontsize=14)
plt.xlabel('Harga (USD)', fontsize=12)
plt.ylabel('Frekuensi', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

Bisa dilihat bahwa, lebih banyak jumlah berlian yang murah pada di *Market* dibandingkan dengan sedikitnya berlian yang mahal.
